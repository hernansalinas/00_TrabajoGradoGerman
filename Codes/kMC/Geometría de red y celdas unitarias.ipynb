```python
import numpy as np
import matplotlib.pyplot as plt

# Par√°metros de la ret√≠cula
a1 = np.array([0.5, 0.7])  # Vector primitivo a1 (en direcci√≥n x)
a2 = np.array([-1, 1])  # Vector primitivo a2 (en direcci√≥n y)
L = 10  # N√∫mero de repeticiones de la celda unitaria en cada direcci√≥n

# Funci√≥n para generar la ret√≠cula a partir de las celdas unitarias
def generate_lattice(a1, a2, L):
    lattice_points = []

    # Generamos las posiciones dentro de la ret√≠cula
    for n1 in range(L):
        for n2 in range(L):
            # Calculamos las posiciones en funci√≥n de los vectores primitivos
            position = n1 * a1 + n2 * a2
            lattice_points.append(position)

    return np.array(lattice_points)

# Generamos la ret√≠cula
lattice_points = generate_lattice(a1, a2, L)

# Visualizaci√≥n de la ret√≠cula
plt.figure(figsize=(6, 6))
plt.scatter(lattice_points[:, 0], lattice_points[:, 1], color='blue', marker='o')

# Agregar las l√≠neas de la celda unitaria
for n1 in range(L):
    for n2 in range(L):
        # Coordenadas de los v√©rtices de la celda unitaria
        origin = n1 * a1 + n2 * a2
        # Dibujar las l√≠neas de la celda unitaria
        plt.plot([origin[0], origin[0] + a1[0]], [origin[1], origin[1] + a1[1]], color='red', lw=1)
        plt.plot([origin[0], origin[0] + a2[0]], [origin[1], origin[1] + a2[1]], color='red', lw=1)

# Configuraci√≥n de la visualizaci√≥n
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.title('Visualizaci√≥n de una Red con Celda Unitaria')
plt.grid(True)
plt.show()
```
